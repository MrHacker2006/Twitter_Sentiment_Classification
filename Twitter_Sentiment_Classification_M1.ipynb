# üßπ 6. Cleaning the dataset

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['Tweet'] = df['Tweet'].str.lower()
```

```python
def clean_tweet(text):
    if pd.isnull(text):
        return ""
    text = re.sub(r"http\S+|www.\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
df['Tweet'] = df['Tweet'].apply(clean_tweet)
```

# üß† 7. Encoding Sentiment

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Sentiment_Encoded'] = label_encoder.fit_transform(df['Sentiment'])
```

```python
df
```

# ‚úÇÔ∏è 8. Applying Train-test split

```python
from sklearn.model_selection import train_test_split
X = df['Tweet']
y = df['Sentiment_Encoded']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# üî§ 9. TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=10000,stop_words='english',ngram_range=(1, 2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

# üå≤ 10. Training Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(X_train_tfidf, y_train)
```

# üß™ 11. Predicting and evaluating the Model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = clf.predict(X_test_tfidf)

print("‚úÖ Accuracy:", round(accuracy_score(y_test, y_pred) * 100, 2), "%\n")
print("üìä Classification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
print("üîÅ Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Example tweet
sample_tweet = "What an amazing sunset today! Nature is truly beautiful üåÖüíõ"


# Step 1: Clean the tweet (use the same function as before)
def clean_tweet(text):
    text = re.sub(r"http\S+|www.\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text.lower()

cleaned_tweet = clean_tweet(sample_tweet)

# Step 2: Vectorize the tweet using the same TF-IDF vectorizer
tweet_vector = vectorizer.transform([cleaned_tweet])

# Step 3: Predict sentiment using the trained model
predicted_label_encoded = clf.predict(tweet_vector)[0]

# Step 4: Decode the numeric prediction back to original label
predicted_label = label_encoder.inverse_transform([predicted_label_encoded])[0]

# Show the result
print("üìù Tweet:", sample_tweet)
print("üìä Predicted Sentiment:", predicted_label)
```
